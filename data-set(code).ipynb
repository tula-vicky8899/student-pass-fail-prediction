```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
n = 100
```

```python
study_hours = np.random.randint(1, 8, n)
attendance = np.random.randint(50, 100, n)
assignments = np.random.randint(3, 11, n)
parent_education = np.random.choice(["low", "medium", "high"], n)
previous_grade = np.random.randint(40, 95, n)
```

```python
final_score = (
    study_hours * 10 + previous_grade * 0.5 + np.random.randint(-5, 6, n)
)
final_score = np.clip(final_score, 0, 100)  # keep within 0-100
```

```python
df = pd.DataFrame({
    "study_hours": study_hours,
    "attendance": attendance,
    "assignments": assignments,
    "parent_education": parent_education,
    "previous_grade": previous_grade,
    "final_score": final_score.astype(int)
})
```

```python
df.to_csv("student_risk_data_large.csv", index=False)
print("Large dataset created successfully!")
```
